```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import OneHotEncoder



data = pd.read_csv("C:/Users/44482978/Desktop/uni/td6/competition_data.csv")
unique_counts_by_column = {}
for column in data.columns:
    unique_counts_by_column[column] = data[column].nunique()    
```

```python
print(unique_counts_by_column["user_id"])
na_count = data["user_id"].isna().sum()
print(na_count)
##deal_print_id tiene casi todo diferente, menos unas 3000 obs, podríamos borrarlo?

```

```python
print(unique_counts_by_column["product_id"])
na_count = data["product_id"].isna().sum()
print(na_count)
##podríamos borarrlo dado que hay tantos NAs?
```

```python
print(unique_counts_by_column["tags"])
na_count = data["tags"].isna().sum()
print(na_count)
unique_values = data["tags"].unique()
print(unique_values)
##one hot codear
```

```python
print(unique_counts_by_column["uid"])
na_count = data["uid"].isna().sum()
print(na_count)
```

```python
encoder = OneHotEncoder(sparse=False)
one_hot_encoded = encoder.fit_transform(data[['platform']])
one_hot_df = pd.DataFrame(one_hot_encoded)
data = pd.concat([data, one_hot_df], axis=1)
data.drop(columns=['platform'], inplace=True)
print(data)
```

```python
columns_to_drop=["etl_version","accepts_mercadopago","site_id","boosted","benefit",]
data.drop(columns=columns_to_drop)
data.drop(columns="date") #esta o print_server_timestamp, la cual tiene + info, pero son redundantes
```
