```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import OneHotEncoder



data = pd.read_csv("C:/Users/44482978/Desktop/uni/td6/competition_data.csv")
unique_counts_by_column = {}
for column in data.columns:
    unique_counts_by_column[column] = data[column].nunique()    
```
